```python
url_base = 'https://npb.jp/bis/yearly/{league}league_{year}.html'
filename_base = 'standings_{year}_{league}.csv'

for i in range(74):
    y = i + 1950
    print(y)
    
    # セリーグ
    url = url_base.format(league='central',year=y)
    lst = pd.read_html(url, flavor='bs4')
    table = lst[3]
    filename = filename_base.format(league='central',year=y)
    table.to_csv('data/' + filename)
    time.sleep(1)
    
    # パリーグ
    url = url_base.format(league='pacific',year=y)
    lst = pd.read_html(url, flavor='bs4')
    table = lst[3]
    filename = filename_base.format(league='pacific',year=y)
    table.to_csv('data/' + filename)
    time.sleep(1)
```
